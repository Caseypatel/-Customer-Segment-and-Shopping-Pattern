# Import

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

import pickle
```

# Load Data

```python
with open('shopping_trends.pkl', 'rb') as f:
       data = pickle.load(f)
```

```python
data.head(2)
```

### Sales by Region

```python
Region_sales = data.groupby('Location')['Purchase Amount (USD)'].agg(['sum', 'min', 'max']).reset_index().sort_values('sum', ascending=True)
```

```python
#visualizing data by season and purchase amount 
seasons = data['Season'].unique()
average_purchase_by_season = data.groupby('Season')['Purchase Amount (USD)'].mean()

plt.figure(figsize=(6, 4))
plt.bar(seasons, average_purchase_by_season)
plt.title("Impact of Season on Purchase")
plt.xlabel("Season")
plt.ylabel("Average Purchase (USD)")
plt.show()
```

Winter and Fall has the high impact on Purchase, that could be due to holiday seasons. 

```python
#visulizing item category by amount purchase 

plt.figure(figsize=(6, 4))
sns.barplot(x='Category', y='Purchase Amount (USD)', data=data, ci=None, palette='viridis')
plt.title("Impact of Category on Purchase")
plt.xticks(rotation=45)
plt.show()
```

The graph shows that the Outerwear Category is a bit lower than the other categories.

```python
#visualizing gender by amount purchased 

gender_purchase = data.groupby('Gender')['Purchase Amount (USD)'].sum()

# Create a donut chart
fig, ax = plt.subplots(figsize=(8, 4))
ax.pie(gender_purchase, labels=gender_purchase.index, autopct='%1.1f%%', startangle=140, colors=['green', 'red'], wedgeprops=dict(width=0.4))
ax.set_title("Impact of Gender on Purchase")
plt.axis('equal')  # Equal aspect ratio ensures that the chart is drawn as a circle.

# Draw a circle in the center to create a donut chart
center_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(center_circle)

plt.show()
```

Male spend (67%) is more than female spend (32%).

```python
#visulizing size by amount purchased 

plt.figure(figsize=(6, 4))
#sns.swarmplot(x='Size', y='Purchase Amount (USD)', data=df, palette='Set2')
sns.barplot(x='Size', y='Purchase Amount (USD)', data=data, palette='viridis')
plt.title("Impact of Size on Purchase")
plt.xlabel('Size')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```

As shown in the chart, the L size had less purchases than the other sizes XL, Small, and Medium.

```python
# visulizing promo code by purchased amount

promo_code= data['Promo Code Used'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 4))
plt.pie(promo_code, labels=promo_code.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral'])
plt.title("Impact of Promo Code Used on Purchase")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

According to pie chart, there seems no impact of Promotion on purchase.

```python
#visulizing location and # of customers
location_counts = data['Location'].value_counts()

location_counts.plot(kind='bar', figsize=(12, 4))
plt.title("Customer Distribution by Location")
plt.xlabel("Location")
plt.ylabel("Number of Customers")
plt.show()
```

As per the chart Montana, California, Idaho, Illinois and Alabama has the higest number of customers

```python
#visualizing dustribution of product categories

category_counts = data['Category'].value_counts()

# Create a figure and axis
plt.figure(figsize=(6, 6))
ax = plt.gca()

# Plot the bar chart with different colors for each bar
bars = plt.bar(category_counts.index, category_counts.values)

# Add labels and title
plt.xlabel('Product Categories')
plt.ylabel('Count')
plt.title('Distribution of Product Categories')
plt.xticks(rotation=90)
```

Clothing is the most popular categories among consumers.

```python
#visualizing top locations by category
top_locations= data['Location'].value_counts().head(5).index


# Create a subplot grid for each location
fig, axes = plt.subplots(5, 1, figsize=(10, 15))

# Iterate through the top locations and create category distribution plots with different colors
for i, location in enumerate(top_locations):
    location_data = data[data['Location'] == location]
    
    # Count the most common product categories in this location
    category_counts = location_data['Category'].value_counts().head(10)
    
    # Create a bar plot for the category distribution with different colors
    ax = axes[i]
    category_counts.plot(kind='bar', ax=ax)
    ax.set_title(f"Categories in {location}")
    ax.set_xlabel("Category")
    ax.set_ylabel("Count")
    ax.set_xticklabels(category_counts.index, rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust subplot layout for a clean appearance
plt.tight_layout()

# Display the visualizations
plt.show()
```

Montana, California, Idaho, Illinois, and Alabama are the top five states in terms of clothing spending.

```python
#Visualizing category distribution by age groups
age_groups = [15, 25, 35, 45, 55, 65]

# Create subplots for each age group
fig, ax = plt.subplots(figsize=(14, 6))

# Create a colormap for age groups
colors = plt.cm.viridis(np.linspace(0, 1, len(age_groups)))

# Initialize a dictionary to store category counts for each age group
category_counts_by_age = {age: [] for age in age_groups}

# Calculate category counts for each age group
for age in age_groups:
    age_group_data = data[(data['Age'] >= age) & (data['Age'] < age + 10)]
    category_counts = age_group_data['Category'].value_counts()
    category_counts_by_age[age] = category_counts

# Create the bar chart
width = 0.15
x = np.arange(len(category_counts_by_age[age_groups[0]].index))

for i, age in enumerate(age_groups):
    category_counts = category_counts_by_age[age]
    ax.bar(x + i * width, category_counts, width=width, label=f'{age}-{age+10}', color=colors[i])

ax.set_xlabel('Product Categories')
ax.set_ylabel('Count')
ax.set_title('Category Distribution by Age Groups')
ax.set_xticks(x + width * (len(age_groups) - 1) / 2)
ax.set_xticklabels(category_counts_by_age[age_groups[0]].index, rotation=45)
ax.legend(title='Age Group')

plt.tight_layout()
plt.show()
```

L size exhibits lower purchases compared to other sizes like XL, Small, and Medium.
             
Promotion does not have a significant impact on purchase behavior.
    
Montana stands out with a remarkable number of customers, indicating a thriving consumer market.
    
Clothing is the most popular product category across all consumers.
    
Accessories are equally popular across age groups, except for those aged 15-25 and 65-75.
    
Footwear is particularly popular among the 45-55 age group.
    
Outerwear enjoys equal popularity across all age groups
